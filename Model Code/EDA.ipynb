Missing Values:

```python
mis_val = [var for var in zillow_df.columns if zillow_df[var].isnull().sum() >0 ]
zillow_df[mis_val].isnull().sum()
```

Relationship between missing values and log error :
Log error is the difference between actual price and zestimate price

```python

# Assuming 'zillow_df' is your DataFrame and 'mis_val' is a list of column names with missing values

def analyze_missing_values(df, var, ax):
    new_df = df.copy()
    new_df[var] = np.where(new_df[var].isnull(), 1, 0)
    new_df.groupby(var)['logerror'].median().plot.bar(ax=ax)
    ax.set_title(var)

# Creating subplots
num_subplots = int(len(mis_val)/4)
fig, axes = plt.subplots(nrows=num_subplots, ncols=4, figsize=(10, 5 * num_subplots))

# Iterating through the variables and plotting on subplots
for i, (var, ax) in enumerate(zip(mis_val, axes.flatten())):
    analyze_missing_values(zillow_df, var, ax)

plt.tight_layout()
plt.show()
```

Cardinality of Categorical variables

```python
for var in cat_vars:
    print(var, len(zillow_df[var].unique())," Categories")
```

```python
def analyze_rare_labels(df ,var ,rare_perc):
    new_df = df.copy()
    tmp = new_df.groupby(var)['logerror'].count() / len(new_df)
    return tmp[tmp < rare_perc]

for var in cat_vars:
    print(analyze_rare_labels(zillow_df, var, 0.01))
```

Continous Variables

```python
cont_vars = [ var for var in num_vars if var not in discrete_vars+year_var]
print('No of continuous variables :',len(cont_vars))
```

```python
# Assuming 'zillow_df' is your DataFrame and 'mis_val' is a list of column names with missing values

def analyze_missing_values(df, var, ax):
    
    new_df = df.copy()
    new_df[var].hist(bins=20,ax=ax)
   
    ax.set_title(var)

# Creating subplots
num_subplots = int(len(cont_vars)/4)
fig, axes = plt.subplots(nrows=num_subplots, ncols=4, figsize=(12, 3 * num_subplots))

# Iterating through the variables and plotting on subplots
for i, (var, ax) in enumerate(zip(cont_vars, axes.flatten())):
    
    analyze_missing_values(zillow_df, var, ax)

plt.tight_layout()
plt.show()
```

In order to convert varriables in to normal distribution , we apply logarithmic transformation of the variables returns that follow a normal distribution

```python
def analyze_log_variation(df, var, ax):
    new_df = df.copy()
    if 0 in new_df[var].unique():
       pass 
    else: 
         new_df[var] = np.log(new_df[var])
         new_df[var].hist(bins=20,ax=ax)
         ax.set_title(var)
     

# Creating subplots
num_subplots = int(len(cont_vars)/4)
fig, axes = plt.subplots(nrows=num_subplots, ncols=4, figsize=(12, 3 * num_subplots))

# Iterating through the variables and plotting on subplots
for i, (var, ax) in enumerate(zip(cont_vars, axes.flatten())):
    
    analyze_log_variation(zillow_df, var, ax)

plt.tight_layout()
plt.show()
```

Analyzing Outliers

```python

def analyze_outliers(df, var, ax):

    new_df = df.copy()
    if 0 in new_df[var].unique():
       pass 
    else: 
         new_df[var] = np.log(new_df[var])
         new_df.boxplot(column=var,ax=ax)
         ax.set_title(var)
     

# Creating subplots
num_subplots = int(len(cont_vars)/4)
fig, axes = plt.subplots(nrows=num_subplots, ncols=4, figsize=(12, 3 * num_subplots))

# Iterating through the variables and plotting on subplots
for i, (var, ax) in enumerate(zip(cont_vars, axes.flatten())):
    
    analyze_outliers(zillow_df, var, ax)

plt.tight_layout()
plt.show()
```
